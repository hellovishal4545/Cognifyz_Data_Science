**Task: Geospatial Analysis**

1)Visualize the locations of restaurants on a
map using latitude and longitude
information.

```python
from shapely.geometry import Point
import geopandas as gpd
from  geopandas import GeoDataFrame
```

```python
#create point geometry from latitude and longitude using shapely

gdf = gpd.GeoDataFrame(df,geometry = gpd.points_from_xy(df.Longitude,df.Latitude))

#create a base map of world using geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

#create a map that fit the screen and plots the restaurants location
gdf.plot(ax = world.plot("continent",legend = True,figsize = (18,15)),marker = 'o',color = 'blue',markersize = 15)
plt.show()
```

2)Analyze the distribution of restaurants
across different cities or countries.

Determine if there is any correlation
between the restaurant's location and its
rating

```python
#distribution of restaurants across diff cities or countries

plt.figure(figsize = (10,6))

sns.countplot(y = "City",data = df,order = df['City'].value_counts().head(10).index,palette= 'Paired')

plt.title("Distribution of Distribution of City")
plt.xlabel("Number of Restaurants")
plt.ylabel("Cities")
plt.show()
```

```python
#correlation between restaurants' location and its rating

plt.figure(figsize = (10,8))

#calculate the correlation between latitude ,logitude and rating
correlation_matrix = df[['Latitude','Longitude','Aggregate rating']].corr()

sns.heatmap(correlation_matrix,annot = True,cmap = 'summer',fmt = '.2f')

plt.title("Correlation between restaurants' location and its rating")
plt.show()
```
