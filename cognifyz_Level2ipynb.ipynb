3) Analyze the availability of online delivery
 among restaurants with different price ranges.


```python
df.head(1)
```

```python
#group by price range and calculate the percentage of restaurands with online delivery
Online_Delivery_by_price_range = df.groupby(['Price range'])['Has Online delivery'].value_counts(normalize=True).unstack()*100
```

```python
Online_Delivery_by_price_range.plot(kind = 'bar',stacked=True,colormap ="viridis",figsize = (10,5))
plt.xlabel('Price range')
plt.ylabel('Percentage of Restaurants with online Delivery')
plt.title('Online Delivery by Price range')
plt.grid()
plt.legend(title = 'Online Delivery',bbox_to_anchor = (1,1))
plt.show()
```

```python
#the restaurants have a online delivery
Online_Delivery_yes=df[df['Has Online delivery'] == 'Yes']

Online_Delivery_counts = Online_Delivery_yes.groupby(['Price range'])['Has Online delivery'].value_counts()
```

```python
Online_Delivery_counts.plot(kind = 'bar',stacked=True,colormap ="summer",figsize = (10,5),label = 'yes')
plt.xlabel('Price range')
plt.ylabel('Number of Restaurants with online Delivery')
plt.title('Online delivery availability by price range')
plt.grid()
plt.legend(title = 'Online Delivery',bbox_to_anchor = (1,1),loc = 'upper left')
plt.xticks(rotation = 0)
plt.show()
```

2) Calculate the average rating for each price
 range.

```python
Avg_Rating_by_price_range = df.groupby(['Price range'])['Aggregate rating'].mean().round(2)
```

```python
print('Average rating for each price range:')
print(Avg_Rating_by_price_range)
```

```python
plt.bar(Avg_Rating_by_price_range.index,Avg_Rating_by_price_range,color = ['skyblue','green','yellow','orange'],width = 0.5)
plt.xlabel('Price range')
plt.ylabel('Average rating')
plt.title('Average rating by price range')
plt.grid()
plt.show()
```
